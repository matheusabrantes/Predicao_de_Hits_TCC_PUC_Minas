```python
# Realizando o left join

def Left_Join_Spotify_Lastfm(df,df2):
    spotify_df = pd.read_csv(df)
    lastfm_df = pd.read_csv(df2)
    combinados = spotify_df.merge(lastfm_df, how='left', left_on=['artist_name'], right_on=['artist_final'])
    combinados.drop_duplicates(subset='uri', keep="first", inplace=True)
    combinados.to_csv(df, index=False)
    combinados

extensao = 'csv'
arquivos = [i for i in glob.glob('*.{}'.format(extensao))]

for f in arquivos:
    loop_df = f
    last_df = r'D:\Documents\Scripts\TCC\Hits\Dataset\Bases Tratadas\Lastfm\FINAL_lastfm.csv'
    Left_Join_Spotify_Lastfm(loop_df,last_df)
```
