```python
# Configurando os CSVs originais, inserindo coluna com o nome do artista principal

def Insere_Artista_Principal(df,df2):
    with open(df,'r', encoding='UTF-8') as csv_entrada:
        with open(df2, 'w', encoding='UTF-8') as csv_saida:
            saida = csv.writer(csv_saida, lineterminator='\n')
            entrada = csv.reader(csv_entrada)

            all = []
            row = next(entrada)
            row.append('artist_name')
            all.append(row)

            for row in entrada:
                row.append(row[1].split('Featuring',1)[0])
                all.append(row)
            saida.writerows(all)

extensao = 'csv'
arquivos = [i for i in glob.glob('*.{}'.format(extensao))]

for f in arquivos:
    entrada_df = f
    saida_df = r'D:\Documents\Scripts\TCC\Hits\Dataset\Bases Tratadas\Spotify\FINAL_'+f

    Insere_Artista_Principal(entrada_df,saida_df)
      
```

```python
# Ajustando os CSVs finais

os.chdir(r"D:\Documents\Scripts\TCC\Hits\Dataset\Bases Tratadas\Spotify")

def Ajusta_Artist_Name(df):
    ajuste_df = pd.read_csv(df)
    ajuste_df['artist_name'] = ajuste_df['artist_name'].str.split('&').str[0]
    ajuste_df['artist_name'] = ajuste_df['artist_name'].str.split(',').str[0]
    ajuste_df['artist_name'] = ajuste_df['artist_name'].str.split('/').str[0]
    ajuste_df['artist_name'] = ajuste_df['artist_name'].str.rstrip(' ')
    ajuste_df['artist_name'] = ajuste_df['artist_name'].str.split(r'(\s)X(\s)').str[0]
    ajuste_df['artist_name'] = ajuste_df['artist_name'].str.split(r'(\s)x(\s)').str[0]
    ajuste_df['artist_name'] = ajuste_df['artist_name'].str.rstrip('')
    ajuste_df['artist_name'] =ajuste_df['artist_name'].replace('Beyonce','Beyoncé')

    ajuste_df.to_csv(df, index=False)

extensao = 'csv'
arquivos = [i for i in glob.glob('*.{}'.format(extensao))]

for f in arquivos:
    df = f

    Ajusta_Artist_Name(df)
```
